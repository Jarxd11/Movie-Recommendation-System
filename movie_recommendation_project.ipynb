```python
import pandas as pd
movies = pd.read_csv("tmdb_5000_movies.csv")
movies.head()
```

```python
credits = pd.read_csv("tmdb_5000_credits.csv")
movies = movies.merge(credits, on='title')
```

```python
movies.shape        # to see rows and columns
movies.columns      # to view all column names
movies.info()       # to check data types and null values
```

```python
movies = movies[['id', 'title', 'overview', 'genres', 'keywords']]
movies.head()
```

```python
movies.dropna(inplace=True)
```

```python
import ast

def safe_convert(obj):
    try:
        L = []
        for i in ast.literal_eval(obj):
            L.append(i['name'])
        return L
    except:
        return []  # return empty list if parsing fails

# Apply safely to both columns
movies['genres'] = movies['genres'].apply(safe_convert)
movies['keywords'] = movies['keywords'].apply(safe_convert)

# Convert overview into list of words
movies['overview'] = movies['overview'].apply(lambda x: x.split() if isinstance(x, str) else [])

# Combine them all safely
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords']

# Join them into one string
movies['tags'] = movies['tags'].apply(lambda x: " ".join(x))

movies.head()

```

```python
def extract_cast(obj):
    try:
        L = []
        for i in ast.literal_eval(obj):
            if len(L) < 3:   # take top 3 actors
                L.append(i['name'])
        return L
    except:
        return []

def extract_director(obj):
    try:
        for i in ast.literal_eval(obj):
            if i['job'] == 'Director':
                return [i['name']]
        return []
    except:
        return []
```

```python
movies['cast'] = movies['cast'].apply(extract_cast)
movies['crew'] = movies['crew'].apply(extract_director)
```

```python
movies.head(3)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies['tags'] = movies['tags'].apply(lambda x: " ".join(x))
```

```python
movies['tags'] = movies['tags'].apply(lambda x: x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(movies['tags']).toarray()

vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
similarity.shape
```

```python
def recommend(movie):
    movie = movie.lower()
    try:
        movie_index = movies[movies['title'].str.lower() == movie].index[0]
    except IndexError:
        print("‚ùå Movie not found in database.")
        return
    
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    print(f"\nüé¨ Movies similar to '{movie.title()}':\n")
    for i in movie_list:
        print(movies.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
